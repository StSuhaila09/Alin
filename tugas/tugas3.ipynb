# Matriks Invers

## Matriks Invers: Pengertian, Sifat, Penerapan, dan Contoh Lebih Lengkap

### Pengertian Matriks Invers

**Matriks invers**, dilambangkan dengan **A⁻¹**, adalah matriks yang ketika dikalikan dengan matriks aslinya **A**, menghasilkan matriks identitas **I**. Matriks identitas adalah matriks persegi dengan elemen diagonal 1 dan elemen lainnya 0.

**Syarat Matriks Memiliki Invers:**

1. **Matriks harus berbentuk persegi** (jumlah baris sama dengan jumlah kolom).
2. **Determinan** matriks **tidak boleh sama dengan 0**.

**Contoh:**

Matriks berikut memiliki invers karena berbentuk persegi dan determinannya tidak sama dengan 0:

```
A = [2 1]
    [3 4]
```

Determinan A = (2)(4) - (1)(3) = 5 ≠ 0

### Sifat-sifat Matriks Invers

1. **Identitas:** A⁻¹A = I = AA⁻¹
2. **Kebalikan:** (A⁻¹)⁻¹ = A
3. **Perkalian:** (AB)⁻¹ = B⁻¹A⁻¹
4. **Konstanta:** (kA⁻¹)⁻¹ = (1/k)A

**Contoh Penerapan Sifat Matriks Invers:**

Misalkan A dan B adalah matriks invers, maka:

- (AB)⁻¹ = B⁻¹A⁻¹
- (2A⁻¹)⁻¹ = (1/2)A

### Mencari Invers dengan Operasi Eliminasi Gauss-Jordan

**Operasi eliminasi Gauss-Jordan** adalah metode untuk mengubah matriks menjadi matriks identitas.

**Langkah-langkah:**

1. Ubah matriks menjadi matriks segitiga atas.
2. Ubah matriks segitiga atas menjadi matriks identitas.
3. Matriks yang sama besar dengan matriks identitas di sisi kanan merupakan invers dari matriks awal.

**Contoh:**

Mencari invers dari matriks A:

```
A = [2 1]
    [3 4]
```

**Langkah 1:**

- Kalikan baris pertama dengan 3/2.
- Kurangkan baris kedua dengan baris pertama.

```
A = [3 1.5]
    [0 1.5]
```

**Langkah 2:**

- Bagi baris kedua dengan 1.5.

```
A = [3 1.5]
    [0 1]
```

**Langkah 3:**

- Kurangkan baris pertama dengan 1.5 kali baris kedua.

```
A = [3 0]
    [0 1]
```

Matriks A sekarang menjadi matriks identitas. Matriks di sisi kanan, yaitu:

```
[1 0]
[0 1]
```

Merupakan invers dari matriks A.

### Solusi Sistem Persamaan Linear dengan Matriks Invers

Misalkan kita memiliki sistem persamaan linear:

```
2x + y = 5
3x + 4y = 7
```

Kita dapat menuliskannya dalam bentuk matriks:

```
[2 1] [x] = [5]
[3 4] [y] = [7]
```

Atau:

```
A X = B
```

Di mana:

- A adalah matriks koefisien
- X adalah matriks variabel
- B adalah matriks konstanta

Solusi X dapat dihitung dengan:

```
X = A⁻¹ B
```

**Contoh:**

Mencari solusi sistem persamaan linear:

```
2x + y = 5
3x + 4y = 7
```

**Langkah 1:**

Mencari invers dari matriks A.

```
A⁻¹ = [1/2 -1/4]
        [-3/2 3/4]
```

**Langkah 2:**

Kalikan invers A dengan matriks B.

```
X = A⁻¹ B = [1/2 -1/4] [5] = [1]
            [-3/2 3/4] [7] = [2]
```

**Solusi:**

x = 1 dan y = 2
