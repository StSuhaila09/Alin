#1. Determinan dari matrik indentitas yg menghasilkan 1

```python
import numpy as np
A= np.array ([[1,0],[0,1]])
determinan = np.linalg.det(A)
print (A,"\n")
print("determinan matrik adalah  identitas ordo 2x2 adalah:", int (determinan))

B = np.array ([[1,0,0],[0,1,0],[0,0,1]])
determinan = np.linalg.det(B)
print (B,"\n")
print("determinan matrik identitas ordo 3x3 adalah:", int (determinan))
```

#2. mendefinisikan sebuah matriks 3x3 dengan elemen-elemennya. Matriks ini kemudian disimpan dalam variabel c dan D

```python
import numpy as np
c= np.array ([[1,2,1],[3,1,2],[2,1,2]])
determinan = np.linalg.det(c)
print (c,"\n")
print("determinan matrik adalah:", int (determinan))

D = np.array ([[3,1,2],[1,2,1],[2,1,2]])
determinan = np.linalg.det(D)
print (D,"\n")
print("determinan matrik jika diubah letak barisnya adalah:", int (determinan))
```

#4. jika terdpat matrik yg menghasilkan determinan "x", dan membuat matrik baru dengan kelipatan"y". maka jika "x" dikaikan "y" maka hasilnya determinan dr matrik baru tsb

```python
import numpy as np
A=  np.array([[3,4],[2,1]])
determinan = np.linalg.det(A)
print (A,"\n")
print("determinan dari matrik A adalah:", int (determinan))

H = np.array ([[6,8],[4,2]])
determinan_H = np.linalg.det(H)
print (H,"\n")
print("determinan dari matrik H adalah:", int(determinan_H.round()))
```
